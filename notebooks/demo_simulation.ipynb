# Streaming Simulation Demo

This notebook demonstrates how to use the `TrajectorySimulator` to simulate a real-time stream of GPS points from a historical CSV trajectory.

```python
import sys
import os
from datetime import datetime

# Ensure src is in pythonpath to import hysoc
project_root = os.path.abspath(os.path.join(os.getcwd(), "..", "src"))
if project_root not in sys.path:
    sys.path.append(project_root)

from hysoc.simulation import TrajectorySimulator
```

## Initialize Simulator
We will load a sample trajectory and configure the simulator to emit points at a faster-than-real-time pace for demonstration.

```python
# Path to a sample file
csv_file = "../data/raw/subset_50/4325685.csv"

# Configuration
interval = 2.0  # Seconds between points
limit = 5       # Stop after n points

print(f"Starting simulation with file: {csv_file}")

sim = TrajectorySimulator(
    file_path=csv_file,
    interval=interval
)

print(f"Simulator initialized for object: {sim._obj_id}")
```

## Run Stream
Iterate through the generator. In a real application, this would likely be an infinite loop or a long-running process.

```python
print(f"Streaming first {limit} points...")

count = 0
start_wall_time = datetime.now()

for point in sim.stream():
    print(f"[{datetime.now().time()}] Emitted: {point}")
    count += 1
    if count >= limit:
        break
        
end_wall_time = datetime.now()
duration = (end_wall_time - start_wall_time).total_seconds()

# Expected duration is (limit - 1) * interval because we don't sleep after the last yield
expected_duration = (limit - 1) * interval

print(f"Finished. {limit} points took {duration:.2f} seconds (expected ~{expected_duration}s)")
```
